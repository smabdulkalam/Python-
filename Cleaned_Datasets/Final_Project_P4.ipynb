```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the files 
df_hindi = pd.read_csv('Cleaned_Movie_Data/cleaned_bollywood_movies.csv')
df_telugu = pd.read_csv('Cleaned_Movie_Data/cleaned_tollywood_movies.csv')
df_indian = pd.read_csv('Cleaned_Movie_Data/cleaned_indian_movies.csv')
```

```python
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
df_all.info()
```

# Analysis 1



```python
best_actress = df_all.groupby('LeadActress')['BoxOffice (Crores)'].mean().sort_values(ascending=False).reset_index()


     
```

```python
best_actress
```

```python
best_actress.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'LeadActress', title = 'Actress and avg collection')
```

# Inferences



Top Performing Actresses: The actresses with the highest average box office collections are likely tied to big-budget, successful films, showcasing their strong fan following and commercial appeal.

Less Representation of Certain Actresses: Some actresses with lower average box office collections may focus on niche or smaller-budget films, which generally have more limited box office success.


# Analysis 2



```python
Profits = df_all.groupby('Budget (Crores)')['BoxOffice (Crores)'].mean().sort_values(ascending=False)
```

```python
Profits_analysis = df_all.groupby('Budget (Crores)').agg(avg_boxoffice=('BoxOffice (Crores)', 'mean')).reset_index()

# Plotting the results
Profits_analysis.plot(x='Budget (Crores)', y='avg_boxoffice', kind='line', title='Budget vs Avg Box Office Collection', xlabel='Budget (Crores)', ylabel='Avg Box Office Collection (Crores)')
```

# Inferences


Higher Budgets Yield Higher Returns: As the budget for a movie increases, there tends to be a corresponding increase in average box office collections, suggesting that bigger investments generally attract larger audiences and more significant revenue.

Diminishing Returns at Very High Budgets: While movies with high budgets tend to perform better, there may be a point where increasing the budget further does not significantly increase box office returns, indicating potential diminishing returns for excessively high-budget films.
