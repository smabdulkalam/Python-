```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the files 
df_hindi = pd.read_csv('Cleaned_Movie_Data/cleaned_bollywood_movies.csv')
df_telugu = pd.read_csv('Cleaned_Movie_Data/cleaned_tollywood_movies.csv')
df_indian = pd.read_csv('Cleaned_Movie_Data/cleaned_indian_movies.csv')
```

```python
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
df_all['BoxOffice (Crores)'].mean()
```

```python
df_all['BoxOffice (Crores)'].plot(kind='hist', bins=10, edgecolor = 'Black', title='Box Office Distribution')


     
```

```python
plt.boxplot(df_all['BoxOffice (Crores)'])
plt.title('BoxOffice (Crores)')
plt.ylabel('Revenue')
plt.show()
```

# Inferences

Histogram Interpretation:
The histogram shows that the majority of movies fall under the Box Office collection bracket of less than ₹300 Crores, indicating this Range is the most common performance level for Indian films.

Box Plot Interpretation:
The boxplot further reinforces this observation — the bulk of movie earnings lie between ₹0 to ₹300 Crores, with the median revenue approximately at ₹296 Crores, as shown by the orange line.
Additionally, the individual data points outside the upper whisker represent outliers, highlighting a few exceptionally high-grossing films that performed significantly above the norm.
