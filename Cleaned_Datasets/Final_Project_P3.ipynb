```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the files 
df_hindi = pd.read_csv('Cleaned_Movie_Data/cleaned_bollywood_movies.csv')
df_telugu = pd.read_csv('Cleaned_Movie_Data/cleaned_tollywood_movies.csv')
df_indian = pd.read_csv('Cleaned_Movie_Data/cleaned_indian_movies.csv')
```

```python
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
avg_genre = round(df_all.groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre
```

```python
avg_genre.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'All Indian movies Genre avg box office')


     
```

# Inferences


Genre Performance Analysis:

-In Indian cinema, the top-performing genres in terms of box office revenue are Biographical Sports Drama, Fantasy, Action, and Science Fiction. These genres tend to have mass appeal, driven by their high entertainment value, visual spectacle, and emotional resonance with a wide audience.

-Conversely, genres like Sports Drama, Sports Action, Thriller, and Family Drama generally see lower box office performance. These often emphasize narrative depth or critical themes, which may appeal to a more niche audience and lack the broad commercial draw of mainstream entertainment genres.


```python
avg_genre_bolly = round(df_hindi[df_hindi['Language'] == 'hindi'].groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre_bolly.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'Bollywood Genre avg box office')

```

# Inferences

The Bollywood film industry appears to have a narrower range of genres, suggesting comparatively less diversity in genre experimentation.

Biographical sports dramas continue to perform well, reflecting a strong audience interest in inspirational, real-life narratives.

Genres such as comedy, action, and romance show higher average box office returns, indicating mainstream appeal and greater mass entertainment value.

In contrast, genres like family drama and thrillers tend to underperform, possibly due to a growing audience preference for more dynamic or historically-rooted storytelling.
